Laden der fMRI-Image-Daten aus Dataframe zu einem BESTIMMTEN ZEITPUNKT:

```python
for index, row in df_img.iterrows():
    print(f"Plotting image for ID: {row['ID']}")
    zeitpunkt_img = row['Bilddaten'].slicer[:, :, :, 0] #Die 4.Dimension gibt den Zeitpunkt an und wird hier mit 0 vordefiniert.
    plotting.plot_epi(zeitpunkt_img, title=row['ID'])
```
